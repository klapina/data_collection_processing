```python
# converting locations to spherical (not necessary for current analysis)
sloc = nyc[['lats','lons']]*np.pi/180. # nyc in spherical
stour = tr[['lats','lons']]*np.pi/180. 
srestr=restr[['lat','lon']]*np.pi/180. 
```

__* Using cross-validation first to optimize bandwidth*__

```python
# FROM http://scikit-learn.org/stable/auto_examples/neighbors/plot_digits_kde_sampling.html
    # use grid search cross-validation to optimize the bandwidth
params = {'bandwidth': np.logspace(-1, 1, 20)}
grid = GridSearchCV(KernelDensity(), params)
```

```python
# include CV http://mark-kay.net/2013/12/24/kernel-density-estimation/
grid = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(1.0e-6, 1.e-3, 30)},
                    cv=5) # 20-fold cross-validation
grid.fit(stour)
print grid.best_params_
grid.fit(sloc)
print grid.best_params_
```

```python
# The best bandwidth (h) is actually around 50m; 2xh approximates a long city block
kde = KernelDensity(bandwidth = 7.85e-6, metric='haversine',kernel='gaussian',algorithm='ball_tree') # 50m 7.85e-6 1
```

```python
#construct KDE of tweets distribution, separately for locals and tourists
kde.fit(sloc)
# evaluate on restaurants LOCAL DENSITY
zloc= np.exp(kde.score_samples(srestr)) 
zloc=zloc 
```

```python
#construct KDE of TOURISTS distribution
kde.fit(stour)
# score_samples() returns the log-likelihood of the samples
ztour = np.exp(kde.score_samples(srestr)) 
ztour=ztour 
```

### Now evaluate local score vs tourists score at each restaurant's location to estimate "locality score"  ###

```python
eps=1.5 
eps1=0.8
restr['dif']=zloc/ztour
is_loc=restr['dif']>eps
is_both1=restr['dif']>eps1 
is_both2= restr['dif']<=eps
is_tour=restr['dif']<eps1
restr_tour=restr[is_tour]
restr_loc=restr[is_loc]
restr_both=restr[is_both1 & is_both2]
```
